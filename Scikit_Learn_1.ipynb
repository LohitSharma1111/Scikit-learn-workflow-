# SCI-KIT Learn workflow

```python
#1 getting  the data ready
import pandas as pd
import numpy as np
heart=pd.read_csv("heart-disease.csv")
heart
```

```python
#creating the X variable(feature matrix)

X=heart.drop("target",axis=1) #independent variable

#creating the Y variable (Labels) 

y=heart["target"] #dependent variable
```

```python
#as this is classification we need to choose the right model and hyperparameters

from sklearn.ensemble import RandomForestClassifier 
clf=RandomForestClassifier(n_estimators=100)
clf.get_params() #to see the parameters
```

```python
#fitting the model to the training data

from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train=train_test_split(X,y , test_size=0.2,)

```

```python
clf.fit(X_train,y_train)
```

```python
#make a predicition
X_test
```

```python
X_train
```

```python
#preds is a conventional name for making the predictions
y_preds=clf.predict(X_test)
y_preds
```

```python
y_test

```

```python
 #evaluating the  model
#to check how well the model is trained (on the train & test data)
clf.score(X_train,y_train) 
```

```python
clf.score(X_test,y_test)   
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_preds))
```

```python
print(classification_report(y_test,y_preds))
```

```python
print(accuracy_score(y_test,y_preds))
```

```python
#5. improving the model 
#trying different n_parameters to check in which prameters does the model givingg the best accuracy rate

np.random.seed(30)
for i in range(10,100,10):
    print(f"trying model with {i} estimators")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print("model accuracy on test data" ,{clf.score(X_test,y_test)})
    print(" ")
    
```

```python
#to save the model 
import pickle 
pickle.dump(clf,open("Loaded_model_1.pk","wb"))
```

```python
load = pickle.load(open("Loaded_model_1.pk","rb"))
load.score(X_test,y_test)
```
